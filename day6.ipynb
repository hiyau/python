```python
import pandas as pd
df = pd.read_csv(r'C:\Users\Hiya\Desktop\logistic_regression_demo.csv')
print(df.head(7))
```

```python
print(df.head())
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(5,4))
plt.scatter(df['Age'],df['Purchased'],c=df['Purchased'],cmap='bwr',label='Purchased')
plt.xlabel('Age')
plt.ylabel('Purchased (0 = No, 1 = Yes)')
plt.title('Hiya Project')
plt.legend(['No purchase','Purchase'])
plt.grid(True)
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
x=df[['Age']]
y=df[['Purchased']]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
model = LogisticRegression()
model.fit(x_train,y_train)
new_age=[[29]]
prediction = model.predict(new_age)
probability = model.predict_proba(new_age)
print(f"Prediction for Age {new_age[0][0]} is : {'Will purchase' if prediction[0]==1 else 'will not purchase'}")
print(f"Probability for Age {new_age[0][0]} is : {probability}")
```
